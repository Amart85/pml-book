## Figure 11.3:<a name='11.3'></a> <a name='leastSquaresGeom'></a>

The target vector $ \bm b  $ is a vector in $\mathbb R ^3$; its orthogonal projection onto the linear subspace is denoted $  \bm b   $. The line from $ \bm b  $ to $  \bm b   $ is the vector of residual errors, whose norm we want to minimize
