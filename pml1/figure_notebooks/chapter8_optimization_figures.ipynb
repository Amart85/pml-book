## Figure 8.6:<a name='8.6'></a> <a name='quadforms'></a> 


 The quadratic form $f( \bm x  ) =  \bm x  ^  \mkern -1.5mu\mathsf T    \mathbf A  \bm x  $ in 2d. (a) $\mathbf A $ is positive definite, so $f$ is convex. (b) $\mathbf A $ is negative definite, so $f$ is concave. (c) $\mathbf A $ is positive semidefinite but singular, so $f$ is convex, but not strictly. Notice the valley of constant height in the middle. (d) $\mathbf A $ is indefinite, so $f$ is neither convex nor concave. The stationary point in the middle of the surface is a saddle point. From Figure 5 of <a href='#Shewchuk1994'>[Jon94]</a>
